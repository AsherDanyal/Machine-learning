```python
df = pd.read_csv("/content/fetal_health.csv")
```

```python
df
```

### **Scalling dataset**

```python
X = df.drop("fetal_health", axis=1)
y = df["fetal_health"]

minmax_scaler = MinMaxScaler()
X_scaled = minmax_scaler.fit_transform(X)
```

```python
X_scaled
```

## **Unsupervised Learing**

```python
silhouette_scores = []
k_range = range(2, 11)  # Silhouette score is undefined for k=1

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    silhouette_scores.append(score)

# Plotting Silhouette Score Curve
plt.figure(figsize=(8, 5))
plt.plot(k_range, silhouette_scores, marker='o', color='green')
plt.title('Silhouette Score vs Number of Clusters (k)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```

### **Unsupervised Learning (Clustering)**

```python
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

print("KMeans Silhouette Score:", silhouette_score(X_scaled, kmeans_labels))
print("KMeans Davies-Bouldin Score:", davies_bouldin_score(X_scaled, kmeans_labels))
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot the clusters
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans_labels, cmap='viridis', s=50, alpha=0.7)
plt.title("KMeans Clustering (k=3) Visualized with PCA")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(scatter, label='Cluster Label')
plt.grid(True)
plt.show()
```

### **Hierarchical Clustering (Agglomerative)**

```python
hierarchical = AgglomerativeClustering(n_clusters=2, linkage='ward')
hierarchical_labels = hierarchical.fit_predict(X_scaled)

print("Hierarchical Silhouette Score:", silhouette_score(X_scaled, hierarchical_labels))
print("Hierarchical Davies-Bouldin Score:", davies_bouldin_score(X_scaled, hierarchical_labels))
```

```python
# Reduce to 2D with PCA for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot the clusters
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=hierarchical_labels, cmap='rainbow', s=50, alpha=0.7)
plt.title("Hierarchical Clustering (Ward linkage) Visualized with PCA")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(scatter, label='Cluster Label')
plt.grid(True)
plt.show()
```

```python
selector = SelectKBest(score_func=chi2, k=10)
X_selected = selector.fit_transform(X_scaled, y)
selected_features = X.columns[selector.get_support()].tolist()
```

```python
selected_features
```

```python
chi2_scores = selector.scores_[selector.get_support()]
percent_contributions = (chi2_scores / chi2_scores.sum()) * 100

# Combine feature names and contributions, then sort by contribution
features_with_scores = list(zip(selected_features, percent_contributions))
features_with_scores.sort(key=lambda x: x[1], reverse=True)

# Unzip sorted features and contributions
sorted_features, sorted_percentages = zip(*features_with_scores)

# Plot
plt.figure(figsize=(10, 6))
bars = plt.bar(sorted_features, sorted_percentages, color='skyblue')
plt.xlabel("Selected Features (sorted)")
plt.ylabel("Contribution (%)")
plt.title("Top 10 Features by Chi-Squared Contribution (Sorted)")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate bars
for bar, pct in zip(bars, sorted_percentages):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f"{pct:.1f}%",
             ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

## **Applying Imbalance Techniques**

```python
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, stratify=y, random_state=42)
```

```python
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
before_counts = Counter(y_train)
after_counts = Counter(y_train_smote)


df_plot = pd.DataFrame({
    'Class': list(before_counts.keys()) * 2,
    'Count': list(before_counts.values()) + list(after_counts.values()),
    'Dataset': ['Before SMOTE'] * len(before_counts) + ['After SMOTE'] * len(after_counts)
})


plt.figure(figsize=(8, 5))
sns.barplot(data=df_plot, x='Class', y='Count', hue='Dataset', palette='Set2')
plt.title('Class Distribution Before and After SMOTE')
plt.xlabel('Class')
plt.ylabel('Number of Samples')
plt.grid(True)
plt.show()
```

## **Supervised Learning models**

### **ADABoost**

```python
from sklearn.ensemble import AdaBoostClassifier

# AdaBoost
ada = AdaBoostClassifier(n_estimators=100, random_state=42)
ada.fit(X_train_smote, y_train_smote)

# Prediction & Evaluation
y_pred_ada = ada.predict(X_test)
print("AdaBoost Accuracy:", accuracy_score(y_test, y_pred_ada))
print("\nClassification Report:\n", classification_report(y_test, y_pred_ada))

cm = confusion_matrix(y_test, y_pred_ada)

# Plotting the confusion matrix using Seaborn
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1', 'Class 2', 'Class 3'], yticklabels=['Class 0', 'Class 1', 'Class 2', 'Class 3'])
plt.title('Confusion Matrix - AdaBoost')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
# Logistic Regression
logreg = LogisticRegression(max_iter=1000, random_state=42)
logreg.fit(X_train_smote, y_train_smote)

# Prediction & Evaluation
y_pred_logreg = logreg.predict(X_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_logreg))
print("\nClassification Report:\n", classification_report(y_test, y_pred_logreg))

# Confusion Matrix
cm_logreg = confusion_matrix(y_test, y_pred_logreg)

# Plotting the confusion matrix using Seaborn
plt.figure(figsize=(6, 5))
sns.heatmap(cm_logreg, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1', 'Class 2', 'Class 3'], yticklabels=['Class 0', 'Class 1', 'Class 2', 'Class 3'])
plt.title('Confusion Matrix - Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

## **Second Dataset**

```python
df=pd.read_csv('Medicaldataset.csv')
```

```python
X = df.drop("Result", axis=1)
y = df["Result"]
```

```python
for col in X:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=5)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(6,4))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance')
plt.grid(True)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
```

```python
print(f"\n Accuracy after PCA feature extraction: {accuracy * 100:.2f}%")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
input_dim = X_scaled.shape[1]
encoding_dim = 5
```

```python
input_layer = Input(shape=(input_dim,))
encoded = Dense(8, activation='relu')(input_layer)
encoded = Dense(encoding_dim, activation='relu')(encoded)
```

```python
decoded = Dense(8, activation='relu')(encoded)
decoded = Dense(input_dim, activation='linear')(decoded)
```

```python
autoencoder = Model(inputs=input_layer, outputs=decoded)
autoencoder.compile(optimizer=Adam(learning_rate=0.01), loss='mse')
```

```python
autoencoder.fit(X_scaled, X_scaled, epochs=100, batch_size=16, shuffle=True, verbose=0)
```

```python
encoder = Model(inputs=input_layer, outputs=encoded)
X_encoded = encoder.predict(X_scaled)
```

```python
X_train_ae, X_test_ae, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
clf_ae = RandomForestClassifier(random_state=42)
clf_ae.fit(X_train_ae, y_train)
y_pred_ae = clf_ae.predict(X_test_ae)
```

```python
accuracy_ae = accuracy_score(y_test, y_pred_ae)
print(f"\n Accuracy with Autoencoder Features: {accuracy_ae * 100:.2f}%")
print(classification_report(y_test, y_pred_ae))
```
